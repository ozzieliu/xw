# Bag of Words!

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer()
count
```

```python
test = df.head(1000)
```

```python
test1 = test[['Answer', 'Clue']]
```

```python
test1['Combined'] = test1[['Answer', 'Clue']].apply(lambda x: ' '.join(x), axis = 1)
```

```python
test1
```

```python
bag = count.fit_transform(test1.Combined)
```

```python
count.get_feature_names()
```

```python
sparse_matrix = pd.DataFrame(bag.toarray(), columns=count.get_feature_names(), index=test1.Combined)
sparse_matrix
```
